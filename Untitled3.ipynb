```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
import matplotlib.pyplot as plt

# Read data from the Excel file
data = pd.read_excel('datatab.xlsx')

# Display the first few rows to verify the data
print(data.head())

# Define the dependent variable
y = data['Salary']    # Dependent variable

# Define the independent variables
# Let's use 'Age', 'Weight', 'Gender', 'Place', 'Company', 'Academic degree' as predictors
X = data[['Age', 'Weight', 'Gender', 'Place', 'Company', 'Academic degree']]

# Handle categorical variables using OneHotEncoding
categorical_features = ['Gender', 'Place', 'Company', 'Academic degree']
numerical_features = ['Age', 'Weight']

# Use ColumnTransformer to apply OneHotEncoder to categorical features
column_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ],
    remainder='passthrough'  # Leave the numerical features as is
)

# Transform the features
X_transformed = column_transformer.fit_transform(X)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X_transformed, y)

# Output the model parameters
print("Model coefficients:")
print(model.coef_)
print(f"Intercept: {model.intercept_:.4f}")

# Make predictions using the model
y_pred = model.predict(X_transformed)

# Calculate the coefficient of determination (R-squared)
r_squared = model.score(X_transformed, y)
print(f"Coefficient of Determination (R^2): {r_squared:.4f}")

# Since we have multiple features, plotting in 2D isn't straightforward.
# However, we can plot the actual vs. predicted salaries.
plt.scatter(y, y_pred, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2)
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.title('Actual vs. Predicted Salary')
plt.show()
```
